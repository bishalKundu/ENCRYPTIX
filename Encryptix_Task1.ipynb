```python
#Importing Libraries

import pandas as pd
from sklearn.linear_model import LogisticRegression as Logreg
from sklearn.preprocessing import LabelEncoder as Le
from sklearn.model_selection import train_test_split as tt
from sklearn.metrics import accuracy_score, confusion_matrix

#Importing Titanic survival dataset

data=pd.read_csv('/content/Titanic-Dataset.csv')
data.head()
```

```python
#Data Count

print(data.count())

# Duplicate value finding from the primary key

print('Duplicate=',data['PassengerId'].duplicated().sum())
```

**There was no duplicate IDs in the primary key so we proceed with the analysis**

```python
#Null Value Finding

data.isna().sum()
```

```python
#Replacing null ages with mean value

data['Age']=data['Age'].fillna(data['Age'].mean())

#New available number of Null vaules

data.isnull().sum()
```

**The null values of age is replaced b the mean age to make the prediction better**

```python
#Count of Embarked category

data['Embarked'].value_counts()
```

```python
#The mode value for Embarked is 'S'(644), Replacing Null with mode

data['Embarked']=data['Embarked'].fillna('S')

#New Null value

data.isnull().sum()
```

**The null values in Embarked variable was replaced with the mode or most repetative port**

```python
#Transforming string to numeric value with LabelEncoder

data['Embarked_Num']=Le().fit_transform(data['Embarked'])
data['Sex_Num']=Le().fit_transform(data['Sex'])
data
```

**The data values for Sex and Embarked was updated to numbers to bring these variables under the predictive Model**

```python
#Removing unnecessary columns

data.drop((['PassengerId','Name','Sex','Embarked','Ticket','Cabin','Fare']), axis=1,inplace=True)
data
```

**Only the necessery columns were kept for further analysis and for building the predictive model**

```python
#Data Information

data.info()
```

```python
# Definig Dependent and Independent variables

x=data.drop(['Survived'],axis=1)
y=data['Survived']
x,y
```

```python
#Training the Logistic model

x_train,x_test,y_train,y_test=tt(x,y,test_size=0.2,random_state=30)
model=Logreg()
model.fit(x_train,y_train)
```

**The model was trained with 20% of the data**

```python
#Predicting value with text data

y_pred=model.predict(x_test)
y_pred
```

```python
#Checking Accuracy and Confusion matrix

print('Accuracy=',accuracy_score(y_test,y_pred))
print('Confusion Matrix=',confusion_matrix(y_test,y_pred))
```

**The accuracy score between the test and prediction data is 77% and the confusion matrix suggests,**

Correctly identified 92 positive cases (True Positives).

Incorrectly identified 16 negative cases as positive (False Positives).

Incorrectly identified 24 positive cases as negative (False Negatives).

Correctly identified 47 negative cases (True Negatives).
